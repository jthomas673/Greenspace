```python
import arcpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os

new_directory = r"C:\Users\jdt45\Documents\Projects\Wiggins Map\File Rasters\ndvi_parcels"

# Change the current working directory
os.chdir(new_directory)
# Set the folder path

current_directory = os.getcwd()

print(current_directory)

files_in_directory = os.listdir('./')

all_files = []
# Print the files
print("Files in current directory:")
for file in files_in_directory:
    all_files.append(file)

# print(all_files)
# Filter the list to include only files of the desired type (e.g., TIF files)
tif_files = [file for file in all_files if file.endswith('.tif')]
print(tif_files)

csv_file = r'values.csv'

with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Longitude', 'Latitude', 'Average_Value'])
```

```python
 # Specify the input TIF file

for file in tif_files:

    # # Read the TIF file as a NumPy array
    raster_array = arcpy.RasterToNumPyArray(file)

    # # Save the NumPy array as a .npy file
    df = pd.DataFrame(raster_array)
#     print(df)
    df[df < 0] = pd.NA
    df_no_zeros = df.replace(0, pd.NA)
    
    avg_value = df_no_zeros.mean().mean()
    print("avgval", avg_value)
    
    desc = arcpy.Describe(file)
    
    # Get the spatial reference
    spatial_ref = desc.spatialReference
    
    # Get the extent
    extent = desc.extent
    
    # Calculate the average location
    avg_x = (extent.XMin + extent.XMax) / 2
    avg_y = (extent.YMin + extent.YMax) / 2
    
    print("geoloc",avg_x,avg_y)
    
    with open(csv_file, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([avg_x, avg_y, avg_value])
```
